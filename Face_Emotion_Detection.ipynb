```python
import cv2
from deepface import DeepFace
import time

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap = cv2.VideoCapture(1)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

cv2.namedWindow('Demo_video')
cv2.moveWindow('Demo_video', 100, 100)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    if frame_count % 10 == 0:
        try:
            small_frame = cv2.resize(frame, (640, 480))
            results = DeepFace.analyze(small_frame, actions=['emotion'])
            dominant_emotion = results[0]['dominant_emotion']

            gray = cv2.cvtColor(small_frame, cv2.COLOR_BGR2GRAY)
            faces = faceCascade.detectMultiScale(gray, 1.1, 4)

            for (x, y, w, h) in faces:
                cv2.rectangle(small_frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                font = cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(small_frame, dominant_emotion, (x, y-10), font, 1, (0, 0, 255), 2, cv2.LINE_4)

            cv2.imshow('Demo_video', small_frame)
        except Exception as e:
            print(f"Error: {e}")

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

    time.sleep(0.1)

cap.release()
cv2.destroyAllWindows()
```
